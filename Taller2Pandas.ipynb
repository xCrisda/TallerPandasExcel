```python
# Importar la librería pandas
import pandas as pd
```

```python
# URL del archivo CSV almacenado en GitHub

url = "https://raw.githubusercontent.com/xCrisda/TallerPandasExcel/refs/heads/main/DataFrame%20personal%20taller%20Pandas.csv"

# Leer el archivo CSV
df = pd.read_csv(url, sep=';')

# Eliminar columnas vacías
df = df.dropna(axis=1, how='all')

# Mostrar las primeras filas
df.head()
```

```python
print(df.shape)
```

# Punto 1 – Creación y Exploración del DataFrame

```python
# Mostrar las primeras 5 filas del DataFrame

primeras_filas = df.head(5)

print("Primeras 5 filas:")
print(primeras_filas)
```

```python
# Mostrar las últimas 10 filas

ultimas_filas = df.tail(10)

print("Últimas 10 filas:")
print(ultimas_filas)
```

```python
# Mostrar el tipo de dato de cada columna

tipos_datos = df.dtypes

print("Tipos de datos:")
print(tipos_datos)
```

```python
# Mostrar cantidad de filas y columnas

filas, columnas = df.shape

print("Cantidad de filas:", filas)
print("Cantidad de columnas:", columnas)
```

```python
# Generar estadísticas generales

estadisticas = df.describe()

print("Descripción estadística:")
print(estadisticas)
```

# Punto 2 – Operaciones Matemáticas entre Columnas

```python
# Crear columna Total_Venta

df["Total_Venta"] = df["Precio"] * df["Cantidad"]

print(df[["Producto", "Total_Venta"]].head())
```

```python
# Crear columna Costo_Total

df["Costo_Total"] = df["Total_Venta"] + df["Costo_envio"]

print(df[["Producto", "Costo_Total"]].head())
```

```python
# Calcular promedio de ventas

promedio_ventas = df["Total_Venta"].mean()

print("Promedio Total_Venta:")
print(promedio_ventas)
```

```python
# Obtener fila con mayor Total_Venta

producto_mayor = df[df["Total_Venta"] == df["Total_Venta"].max()]

print(producto_mayor)
```

```python
# Obtener producto con menor Total_Venta

producto_menor = df[df["Total_Venta"] == df["Total_Venta"].min()]

print(producto_menor)
```

# Punto 3 – Porcentajes y Análisis

```python
# Calcular IVA del 19%

df["IVA"] = df["Total_Venta"] * 0.19

print(df[["Producto", "IVA"]].head())
```

```python
# Ganancia estimada del 25%

df["Ganancia_Estimada"] = df["Total_Venta"] * 0.25

print(df[["Producto", "Ganancia_Estimada"]].head())
```

```python
# Pérdida estimada del 0.05%

df["Perdida_Estimada"] = df["Total_Venta"] * 0.0005

print(df[["Producto", "Perdida_Estimada"]].head())
```

```python
# Calcular porcentaje respecto al total general

total_general = df["Total_Venta"].sum()

df["Porcentaje_Ventas"] = (
    df["Total_Venta"] / total_general
) * 100

print(df[["Producto", "Porcentaje_Ventas"]].head())
```

```python
# Filtrar productos con porcentaje mayor al 15%

productos_15 = df[
    df["Porcentaje_Ventas"] > 15
]

print(productos_15)
```

# Punto 4 – Estadística Descriptiva

```python
# Calcular media, mediana y moda de la columna precio

media_precio = df["Precio"].mean()

mediana_precio = df["Precio"].median()

moda_precio = df["Precio"].mode()

print("Media:", media_precio)

print("Mediana:", mediana_precio)

print("Moda:")
print(moda_precio)
```

```python
# Calcular desviación estándar de la columna cantidad

desviacion = df["Cantidad"].std()

print("Desviación estándar:")
print(desviacion)
```

```python
# Obtener máximo y mínimo de la columna Total_venta

maximo = df["Total_Venta"].max()

minimo = df["Total_Venta"].min()

print("Máximo:", maximo)

print("Mínimo:", minimo)
```

```python
# aqui una explicacion breve de que significa cada medida estadistica obtenida

print("""
Análisis Estadístico:

La media representa el promedio general de los precios.

La mediana representa el valor central de los datos ordenados.

La moda representa el valor que más se repite.

La desviación estándar mide la dispersión de los datos.

El valor máximo representa la mayor venta registrada.

El valor mínimo representa la menor venta registrada.
""")
```

# Punto 5 – Filtrado

```python
# Filtrar productos tecnológicos

productos_tecnologia = df[
    df["Categoria"] == "Tecnología"
]

print(productos_tecnologia)
```

```python
# Ordenar productos por Total_Venta

productos_ordenados = df.sort_values(
    by="Total_Venta",
    ascending=False
)

print(productos_ordenados)
```

```python
# Exportar el DataFrame a un archivo Excel

df.to_excel(
    "resultado_final.xlsx",
    index=False
)

print("Archivo Excel exportado correctamente.")
```
